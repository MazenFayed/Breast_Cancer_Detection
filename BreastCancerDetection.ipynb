#### get features and labels

```python
features = df_new.drop(columns=["diagnosis"],axis=1)
labels = df_new[["diagnosis"]]
```

#### we will split data into train,validate and test data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)
```

#### the training data will be scaled because they have different ranges
#### then we will select features using pca

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
#print(scaled_data)

pca = PCA(n_components=0.99)  # Adjust the number of components according to your needs
X_train = pca.fit_transform(X_train)
print("number of features used = ",X_train.shape[1])

X_train,X_valid,y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

clf = model.fit(X_train,y_train)
```

#### we will validate model before testing

```python
clf.score(X_valid,y_valid)
```

#### we will scale test data according to training data information then we will get accuracy

```python
X_test = scaler.transform(X_test)
X_test = pca.transform(X_test)
clf.score(X_test,y_test)
```
